```python
from pamona import Pamona
import numpy as np
import pamona
```

```python
data2 = np.loadtxt("./scNMT/acc_30.txt")
data3 = np.loadtxt("./scNMT/met_30.txt")
data1 = np.loadtxt("./scNMT/rna_30.txt")
type2 = np.loadtxt("./scNMT/acc_stage.txt")
type3 = np.loadtxt("./scNMT/met_stage.txt")
type1 = np.loadtxt("./scNMT/rna_stage.txt")
type1 = type1.astype(np.int)
type2 = type2.astype(np.int)
type3 = type3.astype(np.int)
data = [data1,data2,data3]
datatype = [type1,type2,type3]
```

```python
Pa = Pamona.Pamona(n_neighbors=40, Lambda=10)
integrated_data, T = Pa.run_Pamona(data)
```

```python
Pa.alignment_score(integrated_data[0], integrated_data[-1])
Pa.test_LabelTA(integrated_data[0],integrated_data[-1],type1,type3)
```

```python
Pa.Visualize([data1,data2,data3], integrated_data, [type1,type2,type3], mode='UMAP') # with datatype
```
