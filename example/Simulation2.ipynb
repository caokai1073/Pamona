```python
from pamona import Pamona
import numpy as np
```

```python
data1 = np.loadtxt("./Simulation2/domain1.txt")
data2 = np.loadtxt("./Simulation2/domain2.txt")
type1 = np.loadtxt("./Simulation2/type1.txt")
type2 = np.loadtxt("./Simulation2/type2.txt")
type1 = type1.astype(np.int)
type2 = type2.astype(np.int)
```

```python
Pa = Pamona.Pamona()
```

```python
integrated_data, T = Pa.run_Pamona([data1, data2])
```

```python
Pa.test_LabelTA(integrated_data[0],integrated_data[-1],type1,type2)
Pa.alignment_score(integrated_data[0], integrated_data[-1][0:200], data2_specific=integrated_data[-1][200:250])
```

```python
Pa.Visualize([data1,data2], integrated_data, mode='UMAP')  # without datatype
Pa.Visualize([data1,data2], integrated_data, [type1,type2], mode='UMAP') # with datatype
```
